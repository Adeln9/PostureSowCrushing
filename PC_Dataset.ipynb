```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import os
```

```python
norm_path = r'D:\PigAbModel\nom_vid'
crush_path = r'D:\PigAbModel\posture_pos_vid'

print(os.listdir(norm_path))
print(os.listdir(crush_path))
```

```python
###dataset structure#####
ndf = pd.DataFrame(columns =['name_','event','inner_seq','side','PC','class'])
cdf = pd.DataFrame(columns =['name_','event','inner_seq','side','PC','class'])


####normal#####
inner_df = pd.DataFrame()
inner_df2 = pd.DataFrame()
for i in os.listdir(norm_path):
    temp1 = os.path.join(norm_path,i)
    
    temp2 = os.path.join(temp1, 'L_PC.csv')
    temp3 = os.path.join(temp1, 'R_PC.csv')
    if os.path.exists(temp2) == True:
        tdf = pd.read_csv(temp2)
        tdf['name_'] = i
        inner_df = pd.concat([inner_df, tdf])
    else:
        None
    
    if os.path.exists(temp3) == True:
        tdf = pd.read_csv(temp3)
        tdf['name_'] = i
        inner_df2 = pd.concat([inner_df2, tdf])
    else:
        None
    
if inner_df.shape == (0,0):
    None
else:
    inner_df.drop(columns = ['Unnamed: 0'], inplace = True)
    inner_df.columns=['inner_seq','PC','name_']
    inner_df['side'] = 'left'
    inner_df['class'] = 'normal'

if inner_df2.shape == (0,0):
    None
else:
    inner_df2.drop(columns = ['Unnamed: 0'], inplace = True)
    inner_df2.columns=['inner_seq','PC','name_']
    inner_df2['side'] = 'right'
    inner_df2['class'] = 'normal'
    
ndf = pd.concat([ndf,inner_df,inner_df2])






####Crush#####

inner_df = pd.DataFrame()
inner_df2 = pd.DataFrame()
for index,i in enumerate(['cam1_event0','cam2_rt_190524','cam1_event3', 'cam1_event4']):
    temp1 = os.path.join(crush_path,i)
   
    for j in os.listdir(temp1):
        temp2 = os.path.join(temp1,j)
        temp3 = os.path.join(temp2, 'L_PC.csv')
        temp4 = os.path.join(temp2, 'R_PC.csv')
        if os.path.exists(temp3) == True:
            tdf = pd.read_csv(temp3)
            tdf['name_'] = j
            tdf['event'] = index
            inner_df = pd.concat([inner_df, tdf])
        else:
            None
            
        if os.path.exists(temp4) == True:
            tdf = pd.read_csv(temp4)
            tdf['name_'] = j
            tdf['event'] = index
            inner_df2 = pd.concat([inner_df2, tdf])
        
        else:
            None
            

if inner_df.shape == (0,0):
    None
else:
    inner_df.drop(columns = ['Unnamed: 0'], inplace = True)
    inner_df.columns=['inner_seq','PC','name_','event']
    inner_df['side'] = 'left'
    inner_df['class'] = 'crush'

if inner_df2.shape == (0,0):
    None
else:
    inner_df2.drop(columns = ['Unnamed: 0'], inplace = True)
    inner_df2.columns=['inner_seq','PC','name_','event']
    inner_df2['side'] = 'right'
    inner_df2['class'] = 'crush'
    
cdf = pd.concat([cdf,inner_df,inner_df2])
```

```python
all_df = pd.concat([ndf,cdf])
```

```python
all_df.to_csv('thresholding_dataset230724.csv')
```
